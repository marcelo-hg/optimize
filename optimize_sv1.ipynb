### criar uma tabela com a comparação da demanda versus alocado

```python
# exemplo
pd.DataFrame(
    [
        ["casaco", "type A", 17, "<valor>"],
        ["casaco", "type C", 32, "<valor>"],
        ["guarda-chuva", "type A", 23, "<valor>"],
    ],
    columns=["product", "type", "demand", "allocated"],
)
```

# Print the results

```python
types = ["b1", "b2", "b3", "b4", "b5"]

dfc = pd.DataFrame()
for product, df in allocation.groupby("product"):
    aux1 = pd.DataFrame()
    for tp, k in zip(types, range(len(types))):
        aux2 = df[tp].iloc[0] * bundles.iloc[k, 1:]
        aux1 = pd.concat([aux1, aux2], axis=1, ignore_index=True)

    aux1 = aux1.sum(axis=1)
    aux1 = aux1.reset_index()
    aux1["product"] = [product] * len(aux1)
    dfc = pd.concat([dfc, aux1], axis=0, ignore_index=True)

dfc.columns = ["type", "allocated", "product"]
aux3 = demand[["product", "type", "demand"]].copy()
dfc = pd.merge(aux3, dfc, on=["product", "type"])
dfc
```
